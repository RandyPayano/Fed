```python
def all_series_tag(tag: str) -> int:
    seriess_result = json.loads(Tags.tags_series(tag)[1])['seriess']
    all_series = {}
    for series in seriess_result[:10]:
        if series:
            all_series[(series['id'])] = series
      
    for series in all_series:
        yield series
    
```
